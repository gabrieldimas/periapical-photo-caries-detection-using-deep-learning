```python
from ultralytics import YOLO

# Load model YOLO untuk klasifikasi
model = YOLO("yolov8s-cls.pt")  # Model klasifikasi pre-trained

# Training model dengan dataset klasifikasi
model.train(
    data="classification_dataset.yaml",  # Path ke dataset YAML
    epochs=30,       # Jumlah epoch
    imgsz=128,       # Ukuran input gambar (default 128x128)
    batch=16         # Batch size
)

# Simpan model dalam format .pt dan .tflite
model.export(format="torch")    # Simpan sebagai .pt
model.export(format="tflite")   # Simpan sebagai .tflite
```
