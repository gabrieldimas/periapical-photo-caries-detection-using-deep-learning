```python
import zipfile
import os

zip_path = "/content/dataset_6.zip"
extract_to = "/content/dataset_6"

os.makedirs(extract_to, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print(f"File berhasil diekstrak ke: {extract_to}")
```

```python
!pip install ultralytics

import os
import shutil

dataset_dir = "/content/dataset_6/dataset_6"

from ultralytics import YOLO

model = YOLO("yolov8n-cls.pt")

model.train(data=dataset_dir, epochs=100, imgsz=640, batch=16, augment=False)
```
